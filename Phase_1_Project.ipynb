```python
fig, ax = plt.subplots(figsize = (20,8))
ax.bar(x = rt_df['studio'], height= rt_df['box_office'])
ax.set_xlabel('studio')
ax.set_ylabel('box_office')
ax.set_title('Relationship between studio and box_office');
```

From the analysis. Paramount Pictures had the highest box office revenue. Microsoft can look at how they produce movies and adapt to their style for their success.

```python
fig, ax = plt.subplots(figsize=(20,8))
plt.barh(rt_df['genre'], width= rt_df['box_office'])
ax.set_title('How the genre of movie perform in the box office')
ax.set_ylabel('genre')
ax.set_xlabel('box office');
```

From the analysis of how the different genres perform at the box office. Comedy had the highest returns. Microsoft should therefore consider creating comedy movies.
